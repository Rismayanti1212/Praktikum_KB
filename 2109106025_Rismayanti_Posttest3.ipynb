## Mengimport Library, Membaca dan Menampilkan Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("healthcare-dataset-stroke-data.csv")
df.info()
```

## Visualisasi Dataset 1 : Grafik Pie
### Berdasarkan Dataset Healthcare Dataset Stroke dengan menggunakan Grafik Pie dapat disimpulkan bahwa :
#### 1. Smokes(Perokok) : Memiliki data paling rendah yaitu sebesar 15.4%
#### 2. Formerly Smoked(Sebelumnya Perokok) : Memiliki data terendah ke-2 yaitu sebesar 17.3%
#### 3. Unknown(Tidak di Ketahui) : Memiliki data terendah ke-3 yaitu sebesar 15.4%
#### 4. Never Smoked(Tidak Pernah Merokok) : Memiliki data tertinggi atau yang paling banyak yaitu sebesar 37.0%

```python
data_pie = df['smoking_status'].value_counts().rename('smoking_status').reset_index(name = 'id_count')
```

```python
plt.pie(data_pie.id_count, 
        labels = data_pie['index'], 
        startangle = 180, 
        autopct = '%.1f%%', 
        explode = (0.1, 0.1, 0, 0),
        shadow = True)
plt.title('Perbandingan Status Smoking Berdasarkan Healthcare Dataset Stroke')
plt.show()
```

## Visualisasi Dataset 2 : Grafik Scatter Plot
### Berdasarkan Dataset Healthcare Dataset Stroke dengan menggunakan Grafik Scatter Plot dapat disimpulkan bahwa :
#### Usia 45 - 75 memiliki rata - rata kadar glukosa yang lebih tinggi dibandingkan Usia 15 - 30 berdasarkan perhitungan BMI namun banyak usia 45 - 75 yang juga memiliki kadar glukosa yang rendah

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,5))
sns.scatterplot(
data = df,
hue = 'age',
x = 'avg_glucose_level',  
y = 'bmi')
plt.title('Grafik Perbandingan Kadar Glukosa pada setiap Usia berdasarkan Indeks Massa Tubuh')
plt.show()
```

## Visualisasi Dataset 3 : Grafik Count Plot
### Berdasarkan Dataset Healthcare Dataset Stroke dengan menggunakan Grafik Count Plot dapat disimpulkan bahwa :
#### 0 atau mengalami Hypertension lebih banyak dialami orang yang tidak pernah merokok namun jika dilihat pada grafik hypertension juga banyak dialami oleh orang yang tidak merokok. 
#### Orang yang merokok juga banyak mengalami hypertension.

```python
sns.set_style("ticks")
plt.figure(figsize = (8,9))
sns.countplot(data = df, x = "hypertension", hue = "smoking_status", palette = "rocket")
plt.title('Perbandingan yang Hipertensi berdasarkan Status Merokok')
```

## Visualisasi Dataset 4 : Grafik Bar Plot
### Berdasarkan Dataset Healthcare Dataset Stroke dengan menggunakan Grafik Bar Plot dapat disimpulkan bahwa :
#### Dari grafik barplot, 0 yang artinya laki-laki, perempuan dan lain-lain yang mungkin saja anak-anak yang mengalami body stroke sedangkan 1 untuk laki-laki dan perempuan tanpa body stroke. Jadi, Jumlah orang tanpa stroke tubuh lebih banyak dibandingkan jumlah orang dengan stroke tubuh. dan dapat dilihat rata-rata orang yang mengalami body stroke adalah usia 20 - 40.

```python
sns.barplot(x='stroke', y='age', hue='gender', data=df)
plt.title('Perbandingan yang Stroke berdasarkan Umur dan Gender')
plt.show()
```
